```python
import pandas as pd
from schedule_client import ScheduleApiClient
```

```python
import os
from dotenv import load_dotenv

load_dotenv()

API_URL = os.getenv("API_URL")

user_street_id = os.getenv("user_street_id")
user_address_id = os.getenv("user_address_id")
```

```python
schedule_client = ScheduleApiClient(API_URL)

streets_json = schedule_client.load_streets()
addrresses_json = schedule_client.load_addresses(streets_json[0]["id"])
schedule_json = schedule_client.load_schedule(addrresses_json[0]["id"])
```

```python
schedule = schedule_json["trashSchedule"]
```

```python
schedule_df = pd.json_normalize(schedule, record_path=['schedule'], meta=['month'])
schedule_df = schedule_df.pivot_table(index='month', columns='type', values='days', aggfunc='first')
schedule_df
```
